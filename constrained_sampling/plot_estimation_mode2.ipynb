```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta
```

```python
folder = 'results/'

ref_beta = np.load(folder + 'AC_Beta_constrained_2mode.npy')
ref_gauss = np.load(folder + 'AC_Gaussian_constrained_2mode.npy')
svgd_teacher = np.load(folder + 'samples_svgd_constrained_teacher_2mode.npy')
svgd_student = np.load(folder + 'samples_svgd_constrained_L2_2mode.npz')

svgd_student_nn = svgd_student['nn_samples']
svgd_student_ss = svgd_student['svgd_samples']
```

```python
ffigs = 'figs/'

num_bins = 50
range_limit = [-1,1]
npoints_plot=500
coords = np.linspace(range_limit[0], range_limit[1], npoints_plot)
fontsize = 12



fig=plt.figure(figsize=(4, 2))
plt.rc('font', size=fontsize)
sns.distplot(ref_beta, num_bins, color="y", kde_kws={"color": "w", "lw": 0} , label='Estimation' )
sns.kdeplot(ref_beta[:,0], bw=.01, color="y")
plt.plot(coords, 0.25*(beta.pdf(coords/2.0 + 0.5, 6, 1.5) + beta.pdf(coords/2.0 + 0.5, 1.5, 6)), 'k-',lw=2.5,  label='Target')

plt.xlim([-1.5,1.5]); plt.ylim([0.0, 1.2])
# plt.legend(fontsize=11, shadow=True, loc=(0.025, 0.6))
fig.savefig(ffigs + 'cons1_beta_2mode.pdf', bbox_inches='tight')

fig=plt.figure(figsize=(4, 2))
plt.rc('font', size=fontsize)
sns.distplot(ref_gauss, num_bins, color="y", kde_kws={"color": "w", "lw": 0})
sns.kdeplot(ref_gauss[:,0], bw=.01, color="y")
plt.plot(coords, 0.25*(beta.pdf(coords/2.0 + 0.5, 6, 1.5) + beta.pdf(coords/2.0 + 0.5, 1.5, 6)), 'k-',lw=2.5,  label='Target')
plt.xlim([-1.5,1.5]); plt.ylim([0.0, 1.2])
fig.savefig(ffigs + 'cons1_gaussian_2mode.pdf', bbox_inches='tight')


fig=plt.figure(figsize=(4, 2))
plt.rc('font', size=fontsize)
sns.distplot(svgd_teacher, num_bins*5, color="y", kde_kws={"color": "w", "lw": 0})
sns.kdeplot(svgd_teacher[:,0], bw=.01, color="y")
plt.plot(coords, 0.25*(beta.pdf(coords/2.0 + 0.5, 6, 1.5) + beta.pdf(coords/2.0 + 0.5, 1.5, 6)), 'k-',lw=2.5,  label='Target')
plt.xlim([-1.5,1.5]); plt.ylim([0.0, 1.2])
fig.savefig(ffigs + 'cons1_svgd_teacher_2mode.pdf', bbox_inches='tight')


fig=plt.figure(figsize=(4, 2))
plt.rc('font', size=fontsize)
sns.distplot(svgd_student_nn, num_bins, color="y", kde_kws={"color": "w", "lw": 0})
sns.kdeplot(svgd_student_nn[:,0], bw=.01, color="y")
plt.plot(coords, 0.25*(beta.pdf(coords/2.0 + 0.5, 6, 1.5) + beta.pdf(coords/2.0 + 0.5, 1.5, 6)), 'k-',lw=2.5,  label='Target')
plt.xlim([-1.5,1.5]); plt.ylim([0.0, 1.2])
fig.savefig(ffigs + 'cons1_svgd_student_2mode.pdf', bbox_inches='tight')


plt.show()
```
