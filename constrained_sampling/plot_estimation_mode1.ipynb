```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta
```

```python
folder = 'results/'

ref_beta = np.load(folder + 'AC_Beta_constrained.npy')
ref_gauss = np.load(folder + 'AC_Gaussian_constrained.npy')
svgd_teacher = np.load(folder + 'samples_svgd_constrained_teacher.npy')
svgd_student = np.load(folder + 'samples_svgd_constrained_L2.npz')

svgd_student_nn = svgd_student['nn_samples']
svgd_student_ss = svgd_student['svgd_samples']
```

```python
ffigs = 'figs/'

num_bins = 50
range_limit = [-1,1]
npoints_plot=100
coords = np.linspace(range_limit[0], range_limit[1], npoints_plot)
fontsize = 12

fig=plt.figure(figsize=(4, 2))
plt.rc('font', size=fontsize)
sns.distplot(ref_beta, num_bins, color="y", label='Estimation' )
plt.plot(coords, beta.pdf(coords/2.0 + 0.5, 4, 2)/2.0, 'k-',lw=2.5,  label='Target' )
plt.xlim([-1.5,1.5]); plt.ylim([0.0, 1.2])
plt.legend(fontsize=11, shadow=True, loc=(0.025, 0.6))
fig.savefig(ffigs + 'cons1_beta.pdf', bbox_inches='tight')

fig=plt.figure(figsize=(4, 2))
plt.rc('font', size=fontsize)
sns.distplot(ref_gauss, num_bins, color="y")
plt.plot(coords, beta.pdf(coords/2.0 + 0.5, 4, 2)/2.0, 'k-',lw=2.5)
plt.xlim([-1.5,1.5]); plt.ylim([0.0, 1.2])
fig.savefig(ffigs + 'cons1_gaussian.pdf', bbox_inches='tight')


fig=plt.figure(figsize=(4, 2))
plt.rc('font', size=fontsize)
sns.distplot(svgd_teacher, num_bins*5, color="y")
plt.plot(coords, beta.pdf(coords/2.0 + 0.5, 4, 2)/2.0, 'k-',lw=2.5)
plt.xlim([-1.5,1.5]); plt.ylim([0.0, 1.2])
fig.savefig(ffigs + 'cons1_svgd_teacher.pdf', bbox_inches='tight')


fig=plt.figure(figsize=(4, 2))
plt.rc('font', size=fontsize)
sns.distplot(svgd_student_nn, num_bins, color="y")
plt.plot(coords, beta.pdf(coords/2.0 + 0.5, 4, 2)/2.0, 'k-',lw=2.5)
plt.xlim([-1.5,1.5]); plt.ylim([0.0, 1.2])
fig.savefig(ffigs + 'cons1_svgd_student.pdf', bbox_inches='tight')

plt.show()
```
